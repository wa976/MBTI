```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab=LabelEncoder()
```

```python
data['Q17']=lab.fit_transform(data['Q17'])
```

```python
data.info()
```

```python
data.head(5)
```

```python
features = ['Q01','Q02','Q03','Q04','Q05','Q06','Q07','Q08','Q09',
            'Q10','Q11','Q12','Q13','Q14','Q15','Q16']
target = ['Q17']

# Split into test and train
X = data[features]
y = data[target].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

parameters = {"max_depth": [3,10,20,30,100,None]}

decision_tree = DecisionTreeClassifier(criterion="gini", min_samples_split = 2)
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(decision_tree, parameters, n_jobs = -1, cv=3, scoring="roc_auc")
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
decision_tree_best = grid_search.best_estimator_
```

```python
y_pred = decision_tree_best.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators = 300, min_samples_split=30, n_jobs =-1)
random_forest.fit(X_train,y_train)
```

```python
grid_search = GridSearchCV(random_forest,parameters, n_jobs = -1, cv=3, scoring="roc_auc")
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
random_forest_best = grid_search.best_estimator_
```

```python
y_pred = random_forest_best.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
answer = [[3,3,3,3,3,3,3,3,3,3,3,3,3,3,3,3,]]
my_predict = random_forest_best.predict(answer)
print(lab.inverse_transform(my_predict))
```
